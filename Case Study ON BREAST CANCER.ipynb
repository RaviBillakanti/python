```python
# Import Data SPSS SAS R
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer(as_frame=True).frame
data
```

```python
# Pandas
data=load_breast_cancer(as_frame=True).frame
data.head()
```

```python
X,Y=data.drop('target',axis=1),data.target
X
```

```python
import matplotlib.pyplot as plt
X.hist()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
x_scl=pd.DataFrame(StandardScaler().fit_transform(X),columns=X.columns)
x_scl
```

```python
from sklearn.decomposition import PCA
pcs=pd.DataFrame(PCA(3).fit_transform(x_scl),columns=['a','b','c'])
pcs
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

dtc=DecisionTreeClassifier()
```

```python
dtc.fit(pcs,Y)
```

```python
plt.figure(dpi=1024)
plot_tree(dtc,filled=True,feature_names=X.columns,class_names=load_breast_cancer().target_names)
plt.show()
```

```python
# Evaluate
dtc.score(pcs,Y)
```
